```python
# Import libraries 

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots we will create

# Read in the data

df = pd.read_csv(r'C:\Users\JFB\Documents\movies\movies.csv')
```

```python
# Looking at the data

df.head()
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data types for our columns

df.dtypes
```

```python
# Replacing missing values with 0
df['budget'] = df['budget'].fillna(0).astype('int64')
df['gross'] = df['gross'].fillna(0).astype('int64')

# Change data type of columns

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
# Create correct Year column

df['yearcorrect'] = df['released'].astype(str).str[0:4]

df.head()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', 20)
```

```python
# Drop any duplicates

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# Scatter plot with budget vs gross

plt.scatter(x=df['budget'],y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Film')

plt.show()
```

```python
df.head()
```

```python
# Plot budget vs gross with Seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
correlation_matrix = df.corr(method='pearson',numeric_only=True)

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# Correlation by Company

df.head()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
correlation_matrix = df_numerized.corr(method='pearson',numeric_only=True)

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized.corr()
```

```python
# Calculate the correlation matrix
correlation_mat = df_numerized.corr()

# Unstack the correlation matrix to view all pairs
corr_pairs = correlation_mat.unstack()

# Set pandas display option to show all rows (no truncation)
pd.set_option('display.max_rows', 20)

# Display the correlation pairs
print(corr_pairs)

# Optionally, reset the display option to the default value (e.g., 60 rows)
pd.reset_option('display.max_rows')
```

```python
high_corr = corr_pairs[(corr_pairs > 0.5) & (corr_pairs.index.get_level_values(0) != corr_pairs.index.get_level_values(1))]

# Display the high correlations
print(high_corr)
```
